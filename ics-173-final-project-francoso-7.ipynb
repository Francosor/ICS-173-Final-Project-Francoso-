```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Sample data
animals = ['Dogs', 'Cats', 'Birds', 'Horses']
counts = [38.4, 25.4, 2.8, 0.7]
```

```python
# Display data through a bar chart
animals = ['Dogs', 'Cats', 'Birds', 'Horses']
counts = [38.4, 25.4, 2.8, 0.7]

fig, ax = plt.subplots()
ax.bar(animals, counts)

for animal, count in zip(animals, counts):
    ax.text(animal, count + 1, str(count), ha='center')
    
ax.set_xlabel('Animal')
ax.set_xlabel('Count')
ax.set_xlabel('Animal Counts Percent of Household Owning')
```

```python
# Create Pie Chart (Percent of Households owning)

fig, ax = plt.subplots()
ax.pie(counts, labels=animals, autopct='%1.1f%%', startangle=90,
      colors=plt.cm.Paired.colors)

# Title
plt.title('U.S. Pet Ownership')

# Legend to the top right
plt.legend(title='Animal Species', loc='lower left')

# Prevent overlapping 
plt.tight_layout()

# Show the plot
plt.show()
```
